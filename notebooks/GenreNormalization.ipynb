```python
import psycopg2
import psycopg2.extras
from pyvis.network import Network
```

```python
conn = psycopg2.connect(user='postgres', database="postgres", password='password', host='localhost', port='5432')
cursor = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
```

```python
cursor.execute('SELECT * FROM genre')
```

```python
results = cursor.fetchall()
colnames = [desc[0] for desc in cursor.description]
```

```python
genres = []
genres_id = {}
for genre in results:
    genres.append(genre[colnames.index('genre')])
    genres_id[genre[colnames.index('genre')]] = genre[colnames.index('genre_id')]
```

```python
categories = {}
genres_assigned = []

keywords_basic = {
    'Pop': ['pop', 'alt z'], 
    'Rock': ['rock'], 
    'Classical': ['classical', 'cello', 'orchestra', 'baroque'], 
    'Jazz': ['jazz', 'bossa nova', 'big band', 'hard bop', 'bebop'], 
    'Blues': ['blues'], 
    'Folk': ['folk'], 
    'Funk': ['funk'], 
    'Lo-fi': ['lo-fi'], 
    'Indie': ['indie', 'alt z'], 
    'Metal': ['metal'], 
    'Hip hop': ['hip hop'], 
    'Edm': ['electronica', 'edm', 'house', 'moog', 'electronic', 'techno', 'vaporwave', 'tronica', 'dubstep', 'trance'],
    'Rap': ['rap'],
    'Soundtrack': ['soundtrack'],
    'Sertanejo': ['sertanejo'],
    'R&b': ['r&b', 'soul'],
    'Punk': ['punk'],
    'Reggae': ['reggae'],
    'Country': ['country'],
    'Instrumental': ['instrumental', 'ambient'],
    'Nacional': ['brasileiro', 'brazilian', 'brasileiros', 'brasileira', 'nacional'],
}

for keyword in keywords_basic.keys():
    categories[keyword] = []
    for genre in genres:
        for key in keywords_basic[keyword]:
            if key in genre:
                genres_assigned.append(genre)
                categories[keyword].append(genre)

print(len(list(set(genres) - set(genres_assigned))))
```

```python
for category in categories:
    for genre in categories[category]:
        query = 'UPDATE genre SET genre_norm = \'' + category + '\' WHERE genre_id = ' + str(genres_id[genre])
        print(query)
```

```python
net = Network(notebook=True)
net.add_node('Genres')
net.add_nodes(list(categories.keys()), size=[15] * len(list(categories.keys())))
for category in categories:
    net.add_edge('Genres', category)
    net.add_nodes(categories[category], size=[5] * len(categories[category]))
    for item in categories[category]:
        net.add_edge(category, item)
```
