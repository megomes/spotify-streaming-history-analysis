```python
import requests

def get_track_bpm_from_deezer(query):
    # 1. Buscar a música
    search_url = "https://api.deezer.com/search"
    search_params = {"q": query, "limit": 1}
    search_resp = requests.get(search_url, params=search_params)
    search_data = search_resp.json()

    if not search_data["data"]:
        print("❌ Nenhum resultado encontrado.")
        return

    # 2. Pega a primeira música
    track_basic = search_data["data"][0]
    track_id = track_basic["id"]

    # 3. Consulta o endpoint detalhado da track
    track_url = f"https://api.deezer.com/track/{track_id}"
    track_resp = requests.get(track_url)
    track = track_resp.json()

    # 4. Exibe dados relevantes
    print(f"\n🎵 {track['title']} - {track['artist']['name']}")
    print(f"🔗 Link: {track['link']}")
    print(f"⏱️ Duração: {track['duration']}s")
    print(f"💓 BPM: {track.get('bpm', 'N/A')}")
    print(f"📶 Gain: {track.get('gain', 'N/A')}")
    print(f"▶️ Preview: {track['preview']}\n")

# Teste
get_track_bpm_from_deezer("concerning the hobbits")
```

```python
import requests

def search_musicbrainz_recording(query):
    print(f"🔎 Buscando no MusicBrainz: {query}")
    url = "https://musicbrainz.org/ws/2/recording/"
    params = {
        "query": query,
        "fmt": "json",
        "limit": 1
    }
    headers = {
    "User-Agent": "MatheusErvilhaSampler/1.0 (matheus@email.com)"
    }
    r = requests.get(url, params=params, headers=headers)
    data = r.json()
    if not data["recordings"]:
        print("❌ Nenhum resultado encontrado no MusicBrainz.")
        return None
    recording = data["recordings"][0]
    title = recording["title"]
    artist = recording["artist-credit"][0]["name"]
    mbid = recording["id"]
    print(f"✅ Encontrado: {title} - {artist} (MBID: {mbid})")
    return mbid

def get_acousticbrainz_features(mbid):
    print(f"🎧 Consultando AcousticBrainz para MBID: {mbid}")
    url = f"https://acousticbrainz.org/api/v1/{mbid}/high-level"
    r = requests.get(url)
    if r.status_code != 200:
        print("❌ Não encontrado no AcousticBrainz.")
        return
    data = r.json()
    highlevel = data.get("highlevel", {})
    bpm = data.get("rhythm", {}).get("bpm", "N/A")
    danceability = highlevel.get("danceability", {}).get("value", "N/A")

    print("\n📈 AcousticBrainz Features:")
    print(f"💓 BPM estimado: {bpm}")
    print(f"💃 Danceability: {danceability}")

query = "Radiohead Creep"
mbid = search_musicbrainz_recording(query)
if mbid:
    get_acousticbrainz_features(mbid)
```
